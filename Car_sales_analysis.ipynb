```
# Import Library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Read File CSV
car_sales = pd.read_csv('Updated_Car_Sales_Data.csv')
```

```
# check the number of rows and columns
print(car_sales.shape)
```

```
# check the first 5 lines
print("5 first line:")
print(car_sales.head())
```

```
# Cek missing value
print(car_sales.isnull().sum())
```

```
# Check for duplicate data
print("Number of rows duplicated:", car_sales.duplicated().sum())
```

```
# Car sales by Transition
sales_by_transmission = car_sales['Transmission'].value_counts()
print(sales_by_transmission)

plt.figure(figsize=(6, 4))
sns.barplot(x=sales_by_transmission.index, y=sales_by_transmission.values, palette='Blues_d')
plt.title('Number of Car Sales Based on Transmission')
plt.xlabel('Transmission Type')
plt.ylabel('Number of Sales')
plt.tight_layout()
plt.show()
```

```
# Car sales by Fuel
sales_by_fuel = car_sales['Fuel Type'].value_counts()
print(sales_by_fuel)

plt.figure(figsize=(7, 4))
sns.barplot(x=sales_by_fuel.index, y=sales_by_fuel.values, palette='Blues_d')
plt.title('Number of Car Sales Based on Fuel')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Sales')
plt.tight_layout()
plt.show()
```

```
# Calculating the number of car sales based on Features
car_sales['Feature_Count'] = car_sales['Options/Features'].fillna('').apply(lambda x: x.count(',') + 1 if x else 0)
sales_by_features = car_sales['Feature_Count'].value_counts().sort_index()
print(sales_by_features)

plt.figure(figsize=(10,6))
sales_by_features.plot(kind='bar', color='skyblue')
plt.title('Number of Car Sales Based on Features')
plt.xlabel('Number of Features')
plt.ylabel('Number of Sales')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```
# Create categories based on price
def categorize_price(price):
    if price <20000:
        return 'Cheap'
    elif price <50000:
        return 'Medium'
    else:
        return 'Expensive'

# Add a price category column
car_sales['Categorize_price'] = car_sales['Price'].apply(categorize_price)
```

```
# Car sales by price category
car_sales['Categorize_price'] = car_sales['Price'].apply(categorize_price)
Number_of_sales_by_price_category = car_sales['Categorize_price'].value_counts()
print(Number_of_sales_by_price_category)

plt.figure(figsize=(8, 5))
sns.barplot(x=Number_of_sales_by_price_category.index, y=Number_of_sales_by_price_category.values, palette='Blues_d')

# Judul dan label
plt.title('Number of Cars Sold by Price Category')
plt.xlabel('Price Category')
plt.ylabel('Number of Sales')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
```

```
# Car sales by Brand
sales_by_brand = car_sales['Car Make'].value_counts()
print(sales_by_brand)

# Take the Top 5 Best Sellers and Bottom 5 Most Popular
top5 = sales_by_brand.head(5)
bottom5 = sales_by_brand.tail(5)

# Merge into one DataFrame
top_bottom_df = pd.concat([top5, bottom5])
top_bottom_df = top_bottom_df.sort_values(ascending=True)


plt.figure(figsize=(10, 6))
sns.barplot(x=top_bottom_df.values, y=top_bottom_df.index, palette='Blues_d')
plt.title('Top 5 & Bottom 5 Car Brands Based on Sales')
plt.xlabel('Number of Sales')
plt.ylabel('Brands')
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```
# Find out which price category the top 5 sales fall into
top5_brand = top5.index.tolist()
df_top5 = car_sales[car_sales['Car Make'].isin(top5_brand)]
df_top5['Categorize_Price'] = df_top5['Price'].apply(categorize_price)

# Crosstab top 5
top5_pivot = pd.crosstab(df_top5['Car Make'], df_top5['Categorize_Price'])

# Plot
top5_pivot.plot(kind='barh', stacked=True, figsize=(10,6), colormap='Blues')
plt.title('Price Category Distribution - Top 5 Car Brands')
plt.xlabel('Number of Sales')
plt.ylabel('Brands')
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```
# Find out which price category the bottom 5 sales fall into
bottom5_brand = bottom5.index.tolist()
df_bottom5 = car_sales[car_sales['Car Make'].isin(bottom5_brand)]
df_bottom5['Categorize_Price'] = df_bottom5['Price'].apply(categorize_price)

# Crosstab bottom 5
bottom5_pivot = pd.crosstab(df_bottom5['Car Make'], df_bottom5['Categorize_Price'])

# Plot
bottom5_pivot.plot(kind='barh', stacked=True, figsize=(10,6), colormap='winter')
plt.title('Price Category Distribution - Bottom 5 Car Brands')
plt.xlabel('Number of Sales')
plt.ylabel('Brands')
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```
# See each feature filled with which price category
pivot_features_to_price = pd.crosstab(car_sales['Feature_Count'], car_sales['Categorize_price'], normalize='index') * 100
print(pivot_features_to_price)

pivot_features_to_price.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='winter')
plt.title('Price Category Composition per Number of Features (%)')
plt.xlabel('Number of Features')
plt.ylabel('Percentage (%)')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```
# See each Fuel type filled with which category
pivot_fuel_vs_price = pd.crosstab(car_sales['Fuel Type'], car_sales['Categorize_price'], normalize='index') * 100
print(pivot_fuel_vs_price)

pivot_fuel_vs_price.plot(kind='bar', stacked=True, figsize=(10,6), colormap='Blues')
plt.title('Price Category Composition per Fuel Type (%)')
plt.xlabel('Fuel Type')
plt.ylabel('Percentage (%)')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
